```python
import pandas as pd
```

```python
pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
#Let's find the mean of beer_servings
```

```python
drinks.beer_servings #You can access each col with dot operator followed by col name
```

```python
drinks.beer_servings.mean() #use mean function to get the mean of the selected series
```

```python
#Let's find the mean of beer_servings in the continent Africa
```

```python
drinks[drinks.continent=='Africa']
```

```python
drinks[drinks.continent=='Africa'].beer_servings
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean() #This is the mean beer_servings for the continent Africa
```

```python
#Let's Assume a task were we need to find the beer_servings for each and every continent separately
```

```python
drinks.groupby('continent')
```

```python
drinks.groupby('continent').beer_servings
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
#Let me represent this graphicaly
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').beer_servings.mean().plot(kind='bar')
```

```python
#Let us find different other parameter using this groupby along with Aggregate
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```python
#We have found the count min max mean of beer_servings for each and every continent
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean']).plot(kind='bar')
```
