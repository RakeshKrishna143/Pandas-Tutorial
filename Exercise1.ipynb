```python
pd.read_csv('olympics.csv')
```

```python
pd.read_csv('olympics.csv',header=1) #header is used to mention the index of the header row
```

```python
pd.read_csv('olympics.csv',header=1,skiprows=5)  #skiprows is used to skip the top rows of our csv
```

```python
pd.read_csv('olympics.csv',header=1,skiprows=[2,5,7,9]) # using skiprows You can also skip a specific row
```

```python
def even_rows(index):
    if index%2==0:
        return True
    return False
pd.read_csv('olympics.csv',header=1,skiprows=lambda x:even_rows(x)) # here I have skipped all the even rows using even_rows function
```

```python
pd.read_csv('olympics.csv',header=1,index_col=0) #index_col is used to specify the index
```

```python
pd.read_csv('olympics.csv',header=1,skipfooter=3) # skipfooter is used to skip the tail rows
```

```python
o = pd.read_csv('olympics.csv',header=1)
o.tail(10) #tail is used to get the last rows of the datas
```
