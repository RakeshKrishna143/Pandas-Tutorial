```python
rating = pd.read_csv('http://bit.ly/imdbratings')
```

```python
rating.title
```

```python
#By following the below two methods we can sort any columns that is the Series
```

```python
rating.title.sort_values() #using sort_values() we can sort any columns by default it works as ascending order
```

```python
rating.title.sort_values(ascending=False) #by setting ascending to False the col is sorted in descending order
```

```python
#Now Let me teach how to sort the DataFrame 
```

```python
rating.sort_values('title') #We have sorted the dataframe by ascending order of the column 'title' 
```

```python
rating.sort_values('duration',ascending=False) #We have sorted the dataFrame in decreasing order of duration
```

```python
rating.sort_values('content_rating')
```

```python
rating.sort_values(['content_rating','duration'])
```

```python
#Here we have sorted the dataframe in content_rating and for each same values of content_rating the datas are sorted in terms of duration
```
